1. Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

	Given List:

	list1 = ["M", "na", "i", "Ra"]
	list2 = ["y", "me", "s", "hul"]

	Output:

	[['M','y'], ['na', me'], ['i', 's'], ['Ra', 'hul']]

```python
list1 = ["M", "na", "i", "Ra"]
list2 = ["y", "me", "s", "hul","ii","oo"]

list3= list(zip(list1,list2))
# add leftover
list3 += [list1[len(list3):] + list2[len(list3):]]
print(list3)

l4=[[a,b] for a,b in zip(list1,list2)]
#add leftover
l4 += list1[len(l4):] + list2[len(l4):]
print(l4)

min_len=min(len(list1),len(list2))
l5=[]
for i in range(min_len):
    l5.append([list1[i],list2[i]])

if len(list1)>min_len:
    l5.append(list1[min_len:])
else:
  l5.append(list2[min_len:])
print(l5)
```

6. Find list of common unique items from two list. and show in increasing order

	Input

	num1 = [23,45,67,78,89,34]
	num2 = [34,89,55,56,39,67]
	Output:

	[34, 67, 89]

```python
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
result=list(set(num1).intersection(set(num2)))
result.sort()
print(result)

common=[]
for i in num1:
  if i in num2:
    if i not in common:
      common.append(i)
common.sort()
print(common)
```

7. Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

	Input:

	['1ac21', '23fg', '456', '098d','1','kls']
	Output:

	['456', '23fg', '1ac21', '1', 'kls', '098d']

```python
list1=['1ac21', '23fg', '456', '098d','1','kls']
def product(s):
  product=1
  for i in s:
    if i.isdigit():
      product *= int(i)
  return product

result=sorted(list1,key=product,reverse=True)
print(result)
```

10. Write a program to find set of common elements in three lists using sets.

			ar2 = [6, 7, 20, 80, 100]

```python
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]
result=set(ar1).intersection(set(ar2)).intersection(set(ar3))
print(list(result))

common=[]
for i in ar1:
  if i in ar2 and i in ar3:
    if i not in common:
      common.append(i)
print(common)
```

16. Write a Python function to check whether a number is perfect or not.¶
	A Perfect number is a number that is half the sum of all of its positive divisors (including itself).

	Example :

	The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6.
	Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6.

	The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.

```python
def perfect(num):
  """
  Funtion to find given number is perfect number or not.
  A Perfect number is a number that is half the sum of all
  of its positive divisors (including itself).
  """
  if type(num)==int:
    if num>0:
      divisor=[]
      for i in range(1,num):
        if num%i==0:
          divisor.append(i)
      sum_divisor=sum(divisor)
      if sum_divisor==num:
        print(num,"is a perfect number")
      else:
        print(num,"is not a perfect number")
    else:
      print("invalid input. Plz enter positive integer only")
  else:
    print("Invalid input. Enter positive integer only")


p1=perfect(6)
p2=perfect(28)
p3=perfect(49)
p4=perfect(88)
p5=perfect([1,2])
```

	Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

18. Write a python function that receives a list of integers and prints out a histogram of bin size 10

	Input:
	[13,42,15,37,22,39,41,50]
	Output:
	{11-20:2,21-30:1,31-40:2,41-50:3}

```python
import math
def histogram(l):
  l1=[13,42,15,37,22,39,41,50]
  min_val=math.floor(min(l1)/10)*10
  max_val=math.ceil(max(l1)/10)*10
  d={}
  for i in range(min_val,max_val,10):
    count=0
    for j in l1:
      if i+1<=j<=i+10:
        count+=1

    d[str(i+1)+'-'+str(i+10)]=count
  return(d)

histogram([13,42,15,37,22,39,41,50])
```
