3. Write a Python program to append text to a file and display the text.

```python
l=['\nsolve the above question','\nsubmit it as soon as possible']
with open('sample.txt','a') as f:
  f.writelines(l)
with open('sample.txt','r') as f:
  print(f.read())
```

4. Write a Python program to read last n lines of a file.

```python
n=int(input('enter no of lines: '))
with open('sample.txt','r') as f:
  lines=f.readlines()
  for i in lines[-n:]:
    print(i,end='')

```

7. Write a python program to find the longest words.

```python
with open('sample.txt','r') as f:
  data=f.read()
  list1=data.split()
  #print(list1)
  max=0
  max_word=''
  for i in list1:
    if len(i)>max:
      max=len(i)
      max_word=i
  print(max_word,max)
```

9. Write a Python program to count the frequency of words in a file.

```python
d={}
with open('sample.txt','r') as f:
  for line in f:
    words=line.split()
    for word in words:
      d[word]=d.get(word,0)+1
print(d)
```

12. Write a Python program to copy the contents of a file to another file .

13. Write a Python program to read a random line from a file.

```python
import random
with open('sample.txt','r') as f:
  lines=f.readlines()
  print(random.choice(lines))
```

14. Write a Python program to assess if a file is closed or not.

```python
f=open('sample.txt','r')
print(f.closed)
f.close()
print(f.closed)

with open('sample.txt','r') as f:
  print(f.closed)  #inside with file is open
print(f.closed)    #outside with file close automatically
```
