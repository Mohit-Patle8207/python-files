1. Write a Python function to find the maximum of three numbers.

```python
#Function to find max of three numbers
def max_of_three(x,y,z):
  if type(x)==type(y)==type(z)==int:
    if x==y==z:
      print("All numbers are equal")
    else:
      if x>=y and x>=z:
        print(x,"is greatest number")
      elif y>=z:
        print(y,"is greatest number")
      else:
        print(z,"is greatest number")
  else:
    print("Invalid input")
```

```python
max_of_three(1,2,3)
max_of_three(2,2,'a')
```

2. Write a Python function to sum all the numbers in a list.

	Sample List : (8, 2, 3, 0, 7)

```python
def sum_list(l):
  sum=0
  for i in l:
    sum+=i
  return sum
sum_list([8,2,3,0,7])
```

6. Write a Python function that accepts a string and counts the number of upper and lower case letters.
	Sample String : 'The quick Brow Fox'
	Expected Output :
	No. of Upper case characters : 3
	No. of Lower case Characters : 12

```python
def lower_upper(str):
  lower=0
  upper=0
  for ch in str:
    if ch.islower():
      lower+=1
    elif ch.isupper():
      upper+=1
  return ("""No. of lower case character is {} No. of upper case character is {}""".format(lower,upper))

lower_upper('The quick Brow Fox')
```

7. Write a Python function that takes a list and returns a new list with distinct elements from the first list.

	Sample List : [1,2,3,3,3,3,4,5]
	Unique List : [1, 2, 3, 4, 5]

```python
def unique(l):
  result=[]
  for i in l:
    if i not in result:
      result.append(i)
  return result

unique([1,2,3,3,3,3,4,5])
```

8. Write a Python function that checks whether a passed string is a palindrome or not.

	Note: A palindrome is a word, phrase, or sequence that reads the same backward as forward, e.g., madam or nurses run.

```python
def palindrome(str):
  rev=str[::-1]
  if rev==str:
    print('{} is a palindrome'.format(str))
  else:
    print('{} is not a palindrome'.format(str))

palindrome('abba')
```

```python
p=lambda x:'palindrome' if x==x[::-1] else 'not a palindrome'
p('madam')
```

9. Write a Python program that accepts a hyphen-separated sequence of words as input and prints the words in a hyphen-separated sequence after sorting them alphabetically.

	Sample Items : green-red-yellow-black-white
	Expected Result : black-green-red-white-yellow

```python
def hyphen_sort(str):
  list1=str.split('-')
  list1.sort()
  result="-".join(list1)
  return result
hyphen_sort('green-red-yellow-black-white')
```
