4. Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.

```python
def input_two_numbers():
  num1=(input("enter first number: "))
  if not num1.isnumeric():
    raise TypeError
  num2=(input("enter second number: "))
  if not num2.isnumeric():
    raise TypeError

try:
  input_two_numbers()
except TypeError:
  print('Takes Numeric value only')
```

```python
#Permission Error occurs when we try to access the locked or restricted file
#here Permission error is raised manually to see how to handel permission error
try:
  with open('sample.txt','w') as f:
    f.write('hi how are you')
    raise PermissionError
except PermissionError:
  print('you dont have permission to access the file')
```
