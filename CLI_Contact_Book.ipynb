Challenge: CLI Contact Book (CSV-Powered)

Create a terminal-based contact book tool that stores and manages contacts using a CSV file.

Your program should:
1. Ask the user to choose one of the following options:
   - Add a new contact
   - View all contacts
   - Search for a contact by name
   - Exit
2. Store contacts in a file called `contacts.csv` with columns:
   - Name
   - Phone
   - Email
3. If the file doesn't exist, create it automatically.
4. Keep the interface clean and clear.

Example:
Add Contact
View All Contacts
Search Contact
Exit

Bonus:
- Format the contact list in a table-like view
- Allow partial match search
- Prevent duplicate names from being added


```python
import csv
data = [
    ["Name", "Phone", "Email"],
    ["Aarav Sharma", "9876543210", "aarav.sharma@example.com"],
    ["Diya Patel", "9123456780", "diya.patel@example.com"],
    ["Rohan Verma", "9988776655", "rohan.verma@example.com"],
    ["Sneha Iyer", "9012345678", "sneha.iyer@example.com"],
    ["Arjun Reddy", "9898989898", "arjun.reddy@example.com"],
    ["Priya Nair", "9765432109", "priya.nair@example.com"],
    ["Karan Mehta", "9345678901", "karan.mehta@example.com"],
    ["Neha Gupta", "9234567890", "neha.gupta@example.com"],
    ["Vikram Singh", "9871234560", "vikram.singh@example.com"],
    ["Ananya Das", "9001122334", "ananya.das@example.com"]
]
with open('contact.csv','w',newline='') as f:
  writer=csv.writer(f)
  writer.writerows(data)

class Contact:
  def __init__(self):
    user_input=int(input("""hi how can i help you
    1. Press 1 to add new contact
    2. Press 2 to view all contact
    3. Press 3 to search contact
    4. Press 4 to exit
    """))

    if user_input == 1:
      self.add_new_contact()

    if user_input == 2:
      self.view_contacts()

    if user_input == 3:
      self.search_contact()

    if user_input == 4:
      return

  def add_new_contact(self):
    cname=input("Enter contact name: ").strip()
    cphone=input("Enter contact phone number: ").strip()
    cemail=input("Enter contact email: ").strip()
    info=[cname,cphone,cemail]
    found=False
    with open('contact.csv', mode='r') as f:
      reader = csv.DictReader(f)
      for row in reader:
        if row['Name'].lower() == info[0].lower():
          found=True
          print("Name already exist")
    if not found:
      with open("contact.csv","a",newline="") as f:
        writer=csv.writer(f)
        writer.writerow(info)
        print(info,' added successfully')

  def view_contacts(self):
    with open("contact.csv", "r") as f:
      reader = csv.reader(f)

      for row in reader:
        print(row)

  def search_contact(self):
    data=input("Enter the name to search: ")
    with open("contact.csv","r") as f:
      reader=csv.DictReader(f)
      for row in reader:
        flag=False
        if row['Name'].lower()==data.lower():
          print(row)
          flag=True
          break
      if not flag:
        print("Contact not found")


```
