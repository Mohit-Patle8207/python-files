 Challenge:  Personal Movie Tracker with JSON

Create a Python CLI tool that lets users maintain their own personal movie database, like a mini IMDb.

Your program should:
1. Store all movie data in a `movies.json` file.
2. Each movie should have:
   - Title
   - Genre
   - Rating (out of 10)
3. Allow the user to:
   - Add a movie
   - View all movies
   - Search movies by title or genre
   - Exit the app

Bonus:
- Prevent duplicate titles from being added
- Format output in a clean table
- Use JSON for reading/writing structured data


```python
import json
with open('movie.json','w') as f:
  json.dump([],f)
class Movies:
  def __init__(self):
    user_input=int(input("""hi how can i help you
    1. Press 1 to add movie
    2. Press 2 to view all movie
    3. Press 3 to search movie
    4. Press 4 to exit
    """))

    if user_input == 1:
      self.add_new_movie()

    if user_input == 2:
      self.view_movie()

    if user_input == 3:
      self.search_movie()

    if user_input == 4:
      return

  def add_new_movie(self):
    title=input("Enter movie title: ").strip()
    genre=input("Enter movie genre: ").strip()
    rating=input("Enter movie rating: ").strip()
    new_movie={"Name":title,"Genre":genre,"Rating":rating}
    found=False
    with open('movie.json', mode='r') as f:
      data = json.load(f)
      for m in data:
        if m['Name'].lower() == new_movie['Name'].lower():
          found=True
          print("Movie already exist")
    if not found:
      with open("movie.json","r") as f:
        data = json.load(f)
      data.append(new_movie)
      with open("movie.json", "w") as f:
        json.dump(data, f, indent=4)
      print('Movie added successfully')

  def view_movie(self):
    with open('movie.json','r') as f:
      data=json.load(f)
    print(data)

  def search_movie(self):
    search=input("Enter movie name or genre to search : ").lower()

    with open('movie.json','r') as f:
      data=json.load(f)

    for movie in data:
      if search in movie['Name'] or search in movie['Genre']:
        print(movie)
```
